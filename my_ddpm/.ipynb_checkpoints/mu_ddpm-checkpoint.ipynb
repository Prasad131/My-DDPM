```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
```

```python
n,T=1000,500
batch_size=128
```

```python
betas = np.linspace(0.001,0.005,T)
alphas = 1-betas
cumprod_alphas = np.cumprod(alphas)
sqrt_cumprod_alphas = np.sqrt(cumprod_alphas)
sqrt_betas = np.sqrt(betas)
sqrt_alphas = np.sqrt(alphas)
sqrt_one_minus_cumprod_alphas = np.sqrt(1-cumprod_alphas)
```

```python
test_size = 10000
x0 = np.random.rand(test_size)
noise = np.random.randn(test_size)
xT = sqrt_cumprod_alphas[T-1]*x0+sqrt_one_minus_cumprod_alphas[T-1]*noise
```

```python
fig,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].hist(x0,bins=25)
ax[0].title.set_text('x0')
ax[1].hist(xT,bins=25)
ax[1].title.set_text('xT')
plt.show()
```

```python
yT = np.random.randn(test_size)
y0 = (yT-sqrt_one_minus_cumprod_alphas[T-1]*noise)/sqrt_cumprod_alphas[T-1]
```

```python
fig,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].hist(y0,bins=25)
ax[0].title.set_text('y0')
ax[1].hist(yT,bins=25)
ax[1].title.set_text('yT')
plt.show()
```
